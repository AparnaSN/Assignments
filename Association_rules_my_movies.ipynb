```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
movies = pd.read_csv("my_movies.csv")
```

```python
movies.head()
```

```python
movies.shape
```

```python
movies.isnull().sum()
```

```python
movies.describe()
```

```python
movies.fillna("0",inplace = True)
movies
```

```python
mymovies=movies.iloc[:,5:]
mymovies
```

# Apriori Algorithm

## Build a model with min support value of 0.11


```python
frequent_itemsets = apriori(mymovies, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending = False)
```

```python
rules[rules.lift>1]
```

```python
rules_new=association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7)
rules_new
```

```python
plt.scatter(rules['support'], rules['confidence'],alpha=0.5)
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
plt.show()
```

```python
plt.scatter(rules_new['support'], rules_new['confidence'],alpha=0.5)
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
plt.show()
```

```python
## Changing a minimum length in apriori algorithm as minimum support as 0.15
```
